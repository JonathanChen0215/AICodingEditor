# AICodingEditor Jupyter Notebook Example
This example uses only Python standard library modules.

```python
import math

rows = []
for x_degree in range(0, 360, 30):
    y1_sin = math.sin(math.radians(x_degree))
    y2_cos = math.cos(math.radians(x_degree))
    rows.append((x_degree, round(y1_sin, 6), round(y2_cos, 6)))

rows
```

```python
for x_degree, y1_sin, y2_cos in rows:
    print(f'AICE_VAR x_degree={x_degree} y1_sin={y1_sin:.6f} y2_cos={y2_cos:.6f}')
```
